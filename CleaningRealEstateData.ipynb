```python
import pandas as pd 
```

```python
data = pd.read_csv('Realestate.csv', low_memory=False)
```

```python
data.head()
```

```python
data.shape
```

```python
# normalize data
data.houseAge.mean() # find column mean
```

```python
houseAge = data.houseAge # create houseAge variabe
norm = (houseAge -houseAge.mean())/houseAge.std()
norm
```

```python
# add normalized data to new column 'houseAgeStandardized'
data['houseAgeStandardized'] = norm 
```

```python
data
```

```python
# drop numberOfConvenienceStores column and apply to df
data.drop('numberOfConvenienceStores', axis=1, inplace=True) 
```

```python
# rename transaction column to transactionDate
data.rename(columns={"transaction": "transactionDate"}, inplace=True)
```

```python
data.head()
```

```python
# display all rows from 0-10 using loc
data.loc[0:10, :] # 11 rows, .loc is inclusive 
```

```python
# use iloc to diplay first 10 rows
data.iloc[0:10, :] # 10 rows, .iloc is exclusive of last number
```

```python
# drop duplicate rows
data.drop_duplicates(inplace=True)
```

```python
data
```

```python
# find any missing values and fill them with the mean for that column
data.fillna(data.mean(), inplace=True)
data
```
