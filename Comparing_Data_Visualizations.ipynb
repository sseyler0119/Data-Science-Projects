```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/sseyler0119/datasets/master/countries.csv')
data.head()
```

### Create a time series that compares life expectancy for Asia and Africa for the year 1952

```python
data_1952 = data[data.year == 1952] # isolate year to 1952
data_1952.head()


```

```python
asia_1952 = data_1952[data_1952.continent== 'Asia'] # isolate Asia from 1952
africa_1952 = data_1952[data_1952.continent== 'Africa'] # isolate Africa from 1952
```

```python
asia_1952.head()
```

```python
africa_1952.head()
```

```python
# find mean and median Life Expectancy for Asia
print('Mean Life Expectancy in Asia:')
print(asia_1952.lifeExpectancy.mean())
print('Median Life Expectancy in Asia:')
print(asia_1952.lifeExpectancy.median())
# find mean and median Life Expectancy for Africa
print('Mean Life Expectancy in Africa:')
print(africa_1952.lifeExpectancy.mean())
print('Median Life Expectancy in Frica:')
print(africa_1952.lifeExpectancy.median())
```

#### Use pyplot's hist() and subplot()

```python
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1) # 2 rows, 1 col, index starts at 1, plt.subplot(211) does the same thing
plt.title('Distribution of Life Expectancy 1952')
plt.hist(asia_1952.lifeExpectancy, edgecolor='black', width=0.95)
plt.ylabel('Asia')
plt.subplot(2,1,2)
plt.hist(africa_1952.lifeExpectancy,edgecolor='black', width=0.95)
plt.ylabel('Africa')
plt.show()
```
