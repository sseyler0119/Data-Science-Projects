```python
import pandas as pd
```

```python
data = pd.read_csv('artwork_data.csv', low_memory=False)
```

```python
data.head()
```

```python
data.filter(items=['id', 'artist']) # just show id and artist columns
```

```python
data.filter(like='year') # show any column with year in it
```

```python
data.filter(like='artist')
```

```python
data.filter(regex="(?i)year") # return all columns with year (case insensitive)
```

```python
data.filter(axis=0, regex="^100.$") # all rows from 1000 to 1009
```

```python
data.filter(axis=0, regex="100(0|2|4|6|8)$") # just return the rows with even values from 1000-1009
```
