```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
fig, ax = plt.subplots(figsize=(10.5, 7.2))
sns.scatterplot(x='total_bill', y='tip', data=tips, color='#f15b2a', ax=ax)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, color='#f15b2a', kind='reg') 
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, color='#f15b2a', kind='hex') 
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, color='#f15b2a', kind='kde', fill=True) #need to add fill=True or rings will be empty
```

```python
#fig = plt.figure(figsize=(21, 14.4))
sns.kdeplot(tips['total_bill'], tips['tip'], fill=True)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(10.5, 7.2))
ax = fig.gca(projection='3d')

ax.plot_trisurf(tips['total_bill'], tips['tip'], tips['size'], color='#f15b2a') # doesn't represent our data well, difficult to see
```

```python
fig, ax = plt.subplots(figsize=(10.5, 7.2))

sns.boxplot(x='time', y='tip', hue='smoker', data=tips, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(10.5, 7.2))

sns.violinplot(x='time', y='tip', hue='smoker', data=tips, ax=ax)
```
