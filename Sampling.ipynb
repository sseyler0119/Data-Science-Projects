```python
data_big = pd.read_csv('https://raw.githubusercontent.com/sseyler0119/datasets/master/obama_too_big.csv', parse_dates=['year_month'])
data = pd.read_csv('https://raw.githubusercontent.com/sseyler0119/datasets/master/obama.csv', parse_dates=['year_month'])
```

```python
plt.figure(figsize=(12, 10))

plt.plot(data_big.year_month, data_big.approve_percent,'o', markersize=2, alpha=0.3)
plt.show() # data is difficult to see, too much data
```

```python
# sample the data
sampled = data_big.sample(n=10000) #10% of data frac=0.1 or n=10000 to specify number of samples
plt.figure(figsize=(12, 10))
plt.plot(sampled.year_month, sampled.approve_percent, 'o', markersize=2, alpha=0.2)
plt.show() # distribution is clusterd around the center of the band
```
