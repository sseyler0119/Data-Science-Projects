```python
import pandas as pd
```

```python
textMining = pd.read_csv('https://raw.githubusercontent.com/sseyler0119/datasets/master/trainTweets.csv')
textMining.head()
```

```python
# Add a column 'word_count' and apply a function that counts how many words are in the column 'tweet'
textMining['word_count'] = textMining['tweet'].apply(lambda x: len(str(x).split(" ")))
# print the first 5 rows
textMining[['tweet', 'word_count']].head()
```

```python
# count the number of characters
textMining['char_count'] = textMining['tweet'].str.len() # this will also account white space
textMining[['tweet', 'char_count']].head()
```

```python
# get the number of special characters used
textMining['hashtags'] = textMining['tweet'].apply(lambda x: len([x for x in x.split()
                                                    if x.startswith('#')]))
textMining[['tweet', 'hashtags']].head()
```

```python
# convert all of the text into lowercase
textMining['tweet'] = textMining['tweet'].apply(lambda x: " ".join(x.lower() for x in x.split()))
textMining['tweet'].head()
```

```python
# remove punctuation
textMining['tweet'] = textMining['tweet'].str.replace('[^\w\s]', '', regex=True) # add regex=True to remove warning
textMining['tweet'].head()
```
