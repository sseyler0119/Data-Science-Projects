```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
# examine the data
diamonds.head()
```

```python
# describe the data
diamonds.describe()
```

Perform univariate analysis

```python
diamonds['carat'].hist(bins=10, figsize=(10,7.5)) 
```

```python
carat = diamonds['carat']
sns.set(rc={"figure.figsize":(10, 7.5)}) # set figure size
p = sns.distplot(carat,
                color='#6bb2e4',
                hist_kws=dict(edgecolor='#356d94', linewidth=2),
                bins=25,
                kde=True,
                kde_kws=dict(shade=True))
```

```python

p = sns.stripplot(x=carat, dodge=True, palette='Blues',
                 jitter = True)
```

```python

p = sns.violinplot(x=diamonds['carat'], inner='box', color='#f15b2a', saturation=0.65)
```

```python
diamonds.drop(['x', 'y', 'z'], axis=1, inplace=True) # view correlation between carat and price, drop x,y,z columns
corrmat = diamonds.corr()
f, ax = plt.subplots(figsize=(10,9))

sns.heatmap(corrmat, ax=ax, cmap="YlGnBu", linewidths=0.2)
```

```python
sns.scatterplot(x="carat", y="price", hue="cut", size="clarity", data=diamonds)
```
