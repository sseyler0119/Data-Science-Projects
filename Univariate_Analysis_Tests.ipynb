```python
import pandas as pd
from scipy import stats
```

Import Iris Dataset

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
                 names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])

df.head()
```

```python
setosa = df[df['class'] == 'Iris-setosa']
setosa.reset_index(inplace=True) # reset index so it starts at 0

setosa.head()
```

```python
versicolor = df[df['class'] == 'Iris-versicolor']
versicolor.reset_index(inplace=True)

versicolor.head()
```

```python
stats.ttest_ind(setosa['sepal_width'], versicolor['sepal_width']) #hypothesis test, # p-value is very small, near 0, so we reject the null hypothesis
```

Perform one-way ANOVA

```python
stats.f_oneway(df['sepal_width'][df['class'] == 'Iris-setosa'],
               df['sepal_width'][df['class'] == 'Iris-versicolor'],
               df['sepal_width'][df['class'] == 'Iris-virginica']) # we reject the null-hypothesis b/c the p-value is near 0, at least one of the sepal_widths must be different than the other classes
```
