```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
mu, sigma = 0, 1

data = np.random.normal(mu, sigma, 1000)
```

```python
p = sns.distplot(data, 
                 color="#F15B2A", 
                 hist_kws=dict(edgecolor="#F15B2A", linewidth=2), 
                 bins=50, 
                 kde=True, 
                 kde_kws=dict(shade=True))
```

```python
p = sns.kdeplot(data,
              color="#F15B2A", 
              shade=True)
```

```python
p = sns.boxplot(data,
                palette=["#F15B2A"])

for e in p.artists:
  r,g,b,a = e.get_facecolor()
  e.set_facecolor((r,g,b,.3))

for i, artist in enumerate(p.artists):
  artist.set_edgecolor("#f15b2a")

for j in range(0,6):
  line = p.lines[j]
  line.set_color("#f15b2a")
  line.set_mfc("#f15b2a")
  line.set_mec("#f15b2a")
```

```python
p = sns.violinplot(data,
                   palette=["#f15b2a"])
```

```python
parts = plt.violinplot(data,
                       vert=False,
                       showmeans=False,
                       showmedians=True)

for p in parts['bodies']:
  p.set_facecolor('#f15b2a')
  p.set_edgecolor("#f15b2a")
  p.set_alpha(0.3)
```

```python
tips = sns.load_dataset('tips')

p = sns.stripplot(x=tips['total_bill'],
                  color = "#f15b2a",
                  jitter = 1)
```

```python
p = sns.swarmplot(x=tips['total_bill'],
                  color = "#f15b2a")
```
