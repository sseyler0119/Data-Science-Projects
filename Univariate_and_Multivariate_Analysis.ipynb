```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
diamonds = sns.load_dataset('diamonds') # load the diamonds dataset from the seaborn library
```

```python
diamonds.head()
```

```python
print(diamonds['cut'].unique()) # display unique values for cut column
```

```python
# Univariate Analysis on 'cut' variable
plt.figure(figsize=(10,10)) # figure size
# display percentage of each unique cut category
plt.pie(diamonds['cut'].value_counts(), startangle=90, autopct='%.1f', labels = ['Ideal', 'Premium', 'Very Good', 'Good', 'Fair'])
```

```python
diamonds.describe() # examine the dataset
```

```python
# Multivariate analysis comparing carat to price categorized by cut and clarity

fig = plt.figure(figsize=(11.5, 8.2)) # figure size

df_sample = diamonds.sample(frac=0.25) # random sample of 25% of the dataset
sns.scatterplot(x = df_sample.carat, y = df_sample.price, hue="cut", size="clarity", data=df_sample)
plt.xlabel('Carats')
plt.ylabel('Price')
plt.title('Carat Vs Price Based on Cut')
```
